# Data analysis exercise

For this exercise, a sample data file is provided which contains
information about "recommendations".  A "recommendation" is the group
of Taskers from which the Client can choose one to book.

The columns are as follows:

- recommendation_id    unique identifier for this recommendation, or set of taskers shown
- timestamp            when this recommendation was shown to the client
- tasker_id            unique identifier for the Tasker
- position             the position of the Tasker in the recommendation set, 1 - first, 2 - second, etc.
- hourly_rate          the hourly rate for the Tasker when they were shown
- num_completed_tasks  the number of Tasks the Tasker had completed in that category, when they were shown
- hired                was the Tasker hired or not?  Only 1 tasker out of a set of recommendations can be hired
- category             the category of work the Client needs help with

### 10. Based on the previous, how would you approach the question of:
### How can we use market data to suggest hourly rates to Taskers that would maximize their opportunity to be hired?


For this question, I think it is appropriate to separately investigate each category for the relationship between hourly rate and hiring chance.

#### For category 'Furniture Assembly':

```python
df_Cat1 = pd.read_csv('/Users/luol12/Desktop/sample.csv')
df_Cat1[df['category']== "Furniture Assembly"]
```

I regard the conversation rate as the chance of a Tasker being hired. 

```python
taskerShowCount_Cat1 = df_Cat1['tasker_id'].value_counts().to_dict()
taskerHiredCount_Cat1 = df_Cat1.groupby('tasker_id')['hired'].apply(lambda x: sum(x.tolist())).to_dict()

tasker_conversationRate_Cat1 = {}
for tasker_id, show_count in taskerShowCount_Cat1.items():
     for taskerID, hired_count in taskerHiredCount_Cat1.items():
            if taskerID == tasker_id:
                tasker_conversationRate_Cat1[taskerID] = hired_count/show_count
tasker_conversationRate_Cat1
```

```python
taskerHourlyRate_Cat1 = df_Cat1.groupby('tasker_id')['hourly_rate'].apply(lambda x: sum(x.tolist())/len(x.tolist())).to_dict()
taskerHourlyRate_Cat1
```

Firstly I would like to look at the scatter plot for hourly rate and conversation rate.

```python
import matplotlib.pyplot as plt
hourlyRate_list = list(taskerHourlyRate_Cat1.values())
conversationRate_list = list(tasker_conversationRate_Cat1.values())
plt.scatter(hourlyRate_list, conversationRate_list)
```

```python
plt.show()
```

The scatter plot indicates that higher hourly rate may lead to lower conversation rate. In category Furniture Assembly, taskers who have relatively high conversation rate (above 0.4) have hourly rate mostly from 25 to 75.

I also take the average position of a tasker into account.

```python
tasker_position = df_Cat1.groupby('tasker_id')['position'].apply(lambda x: sum(x.tolist())/len(x.tolist())).to_dict()
tasker_position
```

```python
ave_position_list = list(tasker_position.values())
plt.scatter(ave_position_list, conversationRate_list)
plt.show()
```

```python
tasker_id_list = list(tasker_position.keys())
tasker_cat1 = pd.DataFrame(
    {'tasker_id': tasker_id_list,
     'ave_position': ave_position_list,
     'hourly_rate': hourlyRate_list,
     'conversation_rate': conversationRate_list
    })
tasker_cat1
```

The multiple regression model describes the response as a weighted sum of the predictors:
conversation_rate = β0 + β1*hourly_rate+ β2*average_positon

```python
import statsmodels.api as sm
X = tasker_cat1[['ave_position', 'hourly_rate']]
y = tasker_cat1['conversation_rate']

## fit a OLS model with intercept on position and hourly rate
X = sm.add_constant(X)
est = sm.OLS(y, X).fit()

est.summary()
```

Unfortunately, from the OLS regression results, the P value indicates that there is no strong evidence to prove the influence of hourly rate and average position on conversation rate.

```python
import statsmodels.api as sm
X = tasker_cat1[['hourly_rate']]
y = tasker_cat1['conversation_rate']

## fit a OLS model with intercept on position and hourly rate
X = sm.add_constant(X)
est = sm.OLS(y, X).fit()

est.summary()
```
